## Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing the dataset

```python
dataset = pd.read_csv("UCI_Credit_Card.csv", index_col = "ID")
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
dataset.info()
```

```python
import seaborn as sns
for col in dataset.columns:
    sns.kdeplot(dataset[col])
    plt.show()
```

```python
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
X
```

```python
y
```

```python
sex_idx = 1          # SEX
education_idx = 2    # EDUCATION
marriage_idx = 3     # MARRIAGE

categorical_indices = [sex_idx, education_idx, marriage_idx]
numerical_indices = [i for i in range(X.shape[1]) if i not in categorical_indices]
```

## Splitting data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## One Hot Encoding and Feature Scaling

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore')
X_train_cat = encoder.fit_transform(X_train[:, categorical_indices])
X_test_cat = encoder.transform(X_test[:, categorical_indices])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_num = scaler.fit_transform(X_train[:, numerical_indices])
X_test_num = scaler.transform(X_test[:, numerical_indices])
```

```python
np.concatenate((X_train_cat, X_train_num), axis=1)
```

## Training the Logistic Regression model

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1))))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_pred, y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

## Training the KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier
for i in range(40, 400, 40):
    clf_knn = KNeighborsClassifier(n_neighbors = i, metric = 'minkowski', p = 2)
    clf_knn.fit(X_train, y_train)
    y_pred_knn = clf_knn.predict(X_test)
    print(f"Accuracy for k = {i}: {accuracy_score(y_pred_knn, y_test)}")
```

```python
clf_knn = KNeighborsClassifier(n_neighbors = 40, metric = 'minkowski', p = 2)
clf_knn.fit(X_train, y_train)
```

```python
y_pred_knn = clf_knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_knn))
```

## Training Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy')
classifier.fit(X_train, y_train)
```

```python
y_pred_dt = classifier.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_dt))
```

```python
from sklearn.ensemble import RandomForestClassifier
for i in range(50, 61):
    classifier_rf = RandomForestClassifier(n_estimators = i, criterion = 'entropy')
    classifier_rf.fit(X_train, y_train)
    y_pred_rf = classifier_rf.predict(X_test)
    print(f"Accuracy for n_estimators = {i}: {accuracy_score(y_pred_rf, y_test)}")
```

```python
classifier_rf = RandomForestClassifier(n_estimators = 55, criterion = 'entropy', class_weight = "balanced")
classifier_rf.fit(X_train, y_train)
```

```python
y_pred_rf = classifier_rf.predict(X_test)
print(classification_report(y_test, y_pred_rf))
cm = confusion_matrix(y_test, y_pred_rf)
print(cm)
accuracy_score(y_pred_rf, y_test)
```
